Now, x are variable that are independent and y is variable that is dependent and we need to predict that y.

Now, we divide on basis of x and y

```python
x = df2.drop('price',axis='columns')
y = df2[['price']]
```

```python
x.head()
```

Now, we divide this data on basis of rows to train ml and test ml.
We divide 80% and 20% data randomly for training and testing purpose.

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train , x_test , y_train , y_test=train_test_split(x,y,test_size=0.2)  # it will split both dataframe where test size is 0.2 means20% and remaining is training size that is 80%
```

Now, training-testing part is done.
Now, we train ml with help of algorithm on data.
For it , we train ml on linear regresssion.

```python
# importing linear regression algorithm class
from sklearn.linear_model import LinearRegression
```

```python
# instance of linear regression algorithm class

lnr = LinearRegression()
```

```python
# training of ml

lnr.fit(x_train,y_train)
```

Now, training of model is done and it's time for testing model or model evaluation,

```python
# checking score of model prediction on training data
lnr.score(x_train,y_train)*100
```

```python
# checking score of model prediction on training data
print("score on test data in % is :",lnr.score(x_test,y_test)*100)
```

Score should be high above 90 and 95.
